```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names() # give all datasets names
```

```python
tips = sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
# sns.set_theme(style="")
# scater plot
sns.relplot(
    data=tips,   # pandas data Frame 
    x="total_bill",
    y="tip",
    # hue -> Splits data into groups using different colors
    hue="smoker",  # all smoker customor showing by different color  
    style="time",  # based on time(lunuch, dinner) point style different
    size="size"
)
```

```python
x = [i+1 for i in range(10)]
y = [i**2 for i in x]

sns.relplot(
    x=x,
    y=y,
    kind="line"
)
```

```python
# Relational Plots - Scatter Plots

tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
# Rleational PLots -> Line plot

sns.lineplot(
    data=flights,
    x="year",
    y="passengers",
    # hue="month",
    # style="month"
)
plt.title("Flights and Passengers")
```

```python
print(tips.head())
```

```python
# Categorical Plots - Bar plot

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
# Categorical Plots - Box plot

# what is tip values over the days
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
    # hue="sex"
)
```

```python
penguins = sns.load_dataset("penguins")
print(penguins.head())
```

```python
# Distribution PLot -> Histogram

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=20,
    kde=True  # smoth curve
)
```

```python
print(flights)
```

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
print(flights_pivot)
```

```python
# Matrix Plots -> Heatmaps
# Heatmaps -> PLot rectangular data as "Color encoded matrix"

sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)

plt.title("Passengers Heatmap")
plt.tight_layout()
```

```python
# Best Practices 
"""
use both:
1. Seaborn -> Individual plot
2. Matplotlib -> For Costomizations
""" 

fig, ax = plt.subplots(2, 2)

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    ax=ax[0][0],
    errorbar=None
)
ax[0][0].set_title("Lineplot: Total Bill by Day and Sex")
ax[0][0].set_xlabel("Days")
ax[0][0].set_ylabel("Total Bill ($)")

# sns.lineplot(
#     data=tips,
#     x="total_bill",
#     y="tip",
#     hue="day",
#     ax=ax[0][1],
# )
# ax[0][1].set_title("Scatter Plot with Seaborn and Matplotlib")
# ax[0][0].set_xlabel("Total Bill ($)")
# ax[0][0].set_ylabel("Tip")

fig.tight_layout()
```

```python
sns.boxplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    ax=ax[0][1],
)
ax[0][1].set_title("Scatter Plot with Seaborn and Matplotlib")
ax[0][0].set_xlabel("Total Bill ($)")
ax[0][0].set_ylabel("Tip")
```
