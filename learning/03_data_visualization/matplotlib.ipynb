```python
# Box Plot
"""
summarizes the distribution of a dataset using five key numbers:
1. Minumum (lowest non-outlier)
2. First Quartile(Q1) - 25th percentile
3. Median(Q2) - 50th percentile
4. First Quartile(Q3) - 75th percentile
5. Maximum (Higher non-outlier)
"""

data = [7, 8, 5, 6, 9, 4, 10, 12, 15]    
# Q2 -> 8  (midium val of sorted data)
# Q1 -> 6 (midium val of 'st to midian')
# Q3 -> 10 (midium val of 'midian to end')

# For minium and maximum
"""
Commponets : Whiskers & Outliers

Upper Whisker (max non-outlier val) : upper = Q3 + (1.5 * IQR)
Q3, Q2, Q1
Lower Whisker (min non-outlier val) : lower = Q1 - (1.5 * IQR)

Whisker length = 1.5   (this is default,  we can change it)
IQR : Inter-Quartile Range
IQR = Q3 - Q1
"""

# Calculating Whiskers:
"""
IQR = 10 - 6 = 4 
lower = 6 - (1.5 * 4) = 6-6 = 0
upper = 10 + (1.5 * 4) = 10+6 = 16

NOTE: not exactly plot the lower & upper val,
we plot those value which is lies in the range of lower and upper

here: lower=0 And upper=16  Not lies in data
"""
# lower = 4   (lowest val of data in range of 0)
# upper = 15  (highest val of data in range of 16)

plt.boxplot(data)
plt.grid()
```

```python
# Subplots

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [np.sqrt(i) for i in x]
y2 = [i*2 for i in x]
y3 = [i**2 for i in x]
y4 = [i**3 for i in x]

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("plot1 - square root")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("plot2 - double")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("plot3 - squar")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("plot4 - cubes")

plt.tight_layout()  # prevent for overlap
```

```python
# till Now 
# all plot are form of - State based API
"""
when we use State based API in MatplotLib
so multiple plot managing going to be very deficult
"""

# So
# when We want to creating multiple plot,
# it i sway better to use - Objecrt Orientend API
```

```python
""" Modern Matplotlib """
# Object Oriented API

fig, axis = plt.subplots(2, 2)
# axis : use for work on induvisual plot
# fig : use for work on ALL plot

axis[0][0].plot(x, y1)
axis[0][0].set_title("plot1 - square root")

axis[0][1].plot(x, y2)
axis[0][1].set_title("plot2 - double")

axis[1][0].plot(x, y3)
axis[1][0].set_title("plot3 - square")

axis[1][1].plot(x, y4)
axis[1][1].set_title("plot4 - cubes")

fig.suptitle("Multiple plots")  # super title
fig.tight_layout()
```

```python

fig, axis = plt.subplots(2, 2, figsize=(7, 7))  # figsize -> size increase

axis[0][0].plot(x, y1)
axis[0][0].set_title("plot1 - square root")
axis[0][0].set_xlabel("x-axis vals")
axis[0][0].set_ylabel("y-axis vals")

axis[0][1].plot(x, y2)
axis[0][1].set_title("plot2 - double")
axis[0][1].set_xlabel("x-axis vals")
axis[0][1].set_ylabel("y-axis vals")

axis[1][0].plot(x, y3)
axis[1][0].set_title("plot3 - square")
axis[1][0].set_xlabel("x-axis vals")
axis[1][0].set_ylabel("y-axis vals")

axis[1][1].plot(x, y4)
axis[1][1].set_title("plot4 - cubes")
axis[1][1].set_xlabel("x-axis vals")
axis[1][1].set_ylabel("y-axis vals")

fig.suptitle("Multiple plots")  # super title
fig.tight_layout()
```
