```python
# Cleaning Data - missing values

df = pd.read_csv("../../data/raw/raw_data.csv")

df.isnull()  # return true for each cell which are NULL(NaN).
df.isna()  # same as isnull
 
df.isnull().sum() # return all number of missing values for each column

df.dropna()  # drop(delete) all rows which have NULL values
df.dropna(axis=1)  # drop all columns which have NULL values

df.fillna(0)  # fill all missing values

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data

df.ffill()  # forword fill
df.bfill()  # backword fill
```

```python
# working on String

df["gender"].str.lower()
df["gender"].str.upper()
df["gender"].str.capitalize()

df["country"].str.strip()
df["name"].str.split()

df["country"].str.contains("India") # whenever contain India show True on that bloack
df["country"].str.contains("india", case=False) 
```

```python
# Transforming Data -> Feature Engineering

df2 =  df.copy()

df2["tax"] = df2["income"].apply(lambda x: "20%" if x >= 60000 else "10%")

gender_map = {"Male" : "M", "Female" : "F", "Unknown" : "U"}
df2["gender"] = df2["gender"].map(gender_map)

df2.assign(new_income = df2["income"] * 1.1)

df2["country"] = df2["country"].replace("USA", "United State")

df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income", "Tax"] # rename all coulmns
# rename for only specific column 
df2.rename(columns = {"Income": "Salary"}) # column rename
df2.rename(index = {1: "First"})  # row rename

df2.sort_values("Income")  # all table sorted in ashending order based on Inocme
df2.sort_values("Income", ascending = False)  # sort in descending order
df2["Income"].sort_values(ascending = False) # only Income column sort and return it

df2["Age"] = df2["Age"].fillna(50)
sorted_df = df2.sort_values(["Income", "Age"])  # sort based on Income , when Income same then sort based on age

sorted_df.sort_index()  # sort index, after sort all table then all value up & down so we use this for sort index

""" --------- reset -------------"""
sorted_df.reset_index()  # it create a new column Index and old column(Index) remain safe with Index column name.
# if we want to drop old index
sorted_df.reset_index(drop=True)

""" --------- rank --------------"""
sorted_df["Ranking"] = sorted_df["Income"].rank()  # 1 is lowest rank 
# if Income ties -> avg rank (lik3: 5 & 6 -> 5.5 & 5.5)
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending = False)  # 1 is Higher rank
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending = False, method="max")  # 1 is Higher rank
# resolving ties case:-
# "dense" -> repeat the rank if ties than go forward
# "min" -> lowset rank store (like: 3 & 4 -> 3 & 3)
# "max" -> maximum rank store (like: 3 & 4 -> 4 & 4)

"""-------- Reorder Columns --------"""
sorted_df[["Name", "Age", "Country", "Gender", "Income", "Ranking", "Tax", "Id"]]
```

```python
""" ---- Grouping Data ----- """
df2.groupby("country")["income"].mean()  # group all data by "country" 
df2.groupby("country")["income"].max()   # and find mean/max/min for each group based on income
df2.groupby("country")["income"].min()  

df2.groupby("gender")["income"].max()
df2.groupby("gender")["income"].min()

df2.groupby("country")["income"].agg(["mean", "max", "min"])  # both are same
df2.groupby("country")["income"].aggregate(["mean", "max", "min"])
# by default column name is method name -> mean, max, min
# but we can also change it
df2.groupby("country")["income"].aggregate(avg_salary="mean", max_salary="max", min_salary="min") 

# if we to applay diff aggregate function to diff. columns with together
df2.groupby("country").agg({
    "income": "mean",
    "age" : "mean"
})
df2.groupby("country").agg({
    "income": "max",
    "age" : "mean"
})

df2.groupby("country").agg(  # change lable column name
    max_salary=("income","max"),
    avg_age=("age","mean")
)
```

```python
"""-- ------- Melt (wide -> long) ---------------"""
df = pd.DataFrame({
    "country" : ["India", "USA", "India", "USA"],
    "year" : [2020, 2020, 2021, 2021],
    "sales" : [110, 100, 90, 120],
    "profit" : [22, 20, 18, 20],
    "tax" : [10, 21, 15, 18]
})

melted_data = df.melt(
    id_vars = ["country", "year"],   # columns to keep
    value_vars = ["sales", "profit", "tax"],  # columns to unpivot
    var_name = "metrics",      # new column name for variable
    value_name = "value"       # new column name for value (default is value)
)

melted_data
```

```python
"""---------- Pivot (long -> wide) ------------"""
original = melted_data.pivot(
    index = ["country", "year"],   # col that become the New row index
    columns = "metrics",   # col whose unique values will be the cols
    values = "value"  # col whose value fill in the 
)

original
```

```python
"""------ Mergeing & Joining-------"""
df_customers = pd.DataFrame({
    "customer_id" : [1, 2, 3, 4],
    "name" : ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id" : [101, 102, 103, 104],
    "customer_id" : [2, 1, 4, 5],
    "amount" : [250, 120, 300, 180]
})

pd.merge(df_customers, df_orders, on="customer_id")  # Inner Join
pd.merge(df_customers, df_orders, on="customer_id", how="left")  # Left Join
pd.merge(df_customers, df_orders, on="customer_id", how="right")  # Right Join
pd.merge(df_customers, df_orders, on="customer_id", how="outer")  # Outer Join
```

```python
"""-------- Concatination -----------"""
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Adam", "Bob", "Charlie"]
})
df2 = pd.DataFrame({
    "id": [4, 5, 6],
    "name": ["David", "Eva", "Frank"]
})

pd.concat([df1, df2]) # Row wise concatination, (df1 top on df2) , index lable remains same
pd.concat([df1, df2], ignore_index = True) # new Index

pd.concat([df1, df2], axis=1) # col wise concatination, (side by side)
```
