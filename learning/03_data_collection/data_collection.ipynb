```python
from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data1.html", "r") as f:
    html_content = f.read()
```

```python
soup = BeautifulSoup(html_content, "lxml")
```

```python

all_h3 = soup.select("h3.country-name")
country_name = []
for h3 in all_h3:
    country_name.append(h3.get_text(strip=True))
```

```python
country_population = []
all_span = soup.select("span.country-population")
for span in all_span:
    country_population.append(int(span.get_text(strip=True)))
```

```python
import pandas as pd
df = pd.DataFrame({
    "country": country_name,
    "population": country_population
})
```

```python
with open("cleaned_data/data1.csv", "w") as f:
    df.to_csv(f)
```

```python
# Assignment ---
import requests
URL = "https://quotes.toscrape.com/page/"
```

```python
pageNo = 1
while(True):
    try:
        res = requests.get(URL+str(pageNo), timeout=(5, 10))
        res.raise_for_status()
        
        soup = BeautifulSoup(res.text, "lxml")
        if(soup.select("div.quote") == []):   # no more quote available on site
            break

        with open("scraped_data/data2_pageNo"+str(pageNo)+".html", "w", encoding="utf-8") as f:
            f.write(res.text)
    except requests.exceptions.RequestException as e:
        print("Error: ", e)
    pageNo += 1
```

```python
quotes = []
authors = []
for i in range(1, 11):
    with open("scraped_data/data2_pageNo"+str(i)+".html", "r", encoding="utf-8") as f:
        html_content = f.read()
    soup = BeautifulSoup(html_content, "lxml")
    all_div_quote = soup.select("div.quote")
    for div_quote in all_div_quote:
        for tag in div_quote.select("div.tags")[0].select("a.tag"):
            if tag.get_text() == "life":
                quotes.append(div_quote.select("span.text")[0].get_text())
                authors.append(div_quote.select("small.author")[0].get_text())
            break
```

```python
import pandas as pd
df  = pd.DataFrame({
    "quote": quotes,
    "author": authors
})
```

```python
with open("cleaned_data/date2.csv", "w", encoding="utf-8") as f:
    df.to_csv(f)
```
