```python
import torch
import torch.nn as nn  # building blocks of nn used for activation function or creating layers
```

```python
# input
# tenser -> is a multi-dim array
"""
0D -> scaler (const vals)
1D -> vector
2D -> matrix
3D+ -> tensors (used in images and videos store)
"""
torch.manual_seed(42)  # used for same output evrytime as random_state
inputs = torch.tensor([1.0, 2.0, 3.0])
```

```python
# creating neuron
# Linear Activation funx -> calculate the weighted sum
neuron = nn.Linear(in_features=3, out_features=1)
```

```python
# output
output = neuron(inputs)
```

```python
output
```

```python
# verify  ... formula
neuron.weight @ inputs + neuron.bias
```
