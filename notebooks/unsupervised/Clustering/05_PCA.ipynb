# **PCA** - Principle Components Analysis (Dimensionality Reduction) 
- here we will work on **iris dataset**
- iris data set have 4 features (4D)
- So we reduce it diamension: **4D->2D**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()

X = pd.DataFrame(iris.data)
y = iris.target
```

```python
X.head()
```

## Step 1. Standardize the Data 

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### PCA code Structure: 

class **sklearn.decomposition.PCA**(
`n_components`=None, *, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', n_oversamples=10, power_iteration_normalizer='auto', random_state=None
)

**n_components** :
    - Number of components(dimension) we want to keep.
    - if n_components is not set (none)
        `n_components= **min**(n_samples, n_features)`

```python
# PCA

from sklearn.decomposition import PCA

pca = PCA(
    n_components=2,
)

X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
print("Explained variance ratio: ", pca.explained_variance_ratio_)
# pc1 and pc2 variance(captures) value
```

```python
print(pca.components_)  # eigent vectors of pc1 and pc2
```

```python
# visualize
plt.figure(figsize=(8,6))

plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=y
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA")
```
