# Anomalies Detection - by **DBSCAN** 

```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(
    n_samples=500,
    noise=0.1,
    random_state=42
)
```

```python
# Scalling
scaler = StandardScaler()
X_scaled =  scaler.fit_transform(X)
```

```python
# visualize
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], c=y)
```

```python
X.shape
```

## Note:
- ### `min_samples` = Optimized value
    - **too low** -> noise becomes normal samples
    - **too high** -> normal data becomes noise 
    - **~****`2x no. of features`** -> general rule of thumbs for set min_samples value

- ### **`eps`** -> Epsilon ($\epsilon$) value
     - its define the **size** of **$\epsilon$-negighborhood**
     - general convention: (as beginner use)
       - low dim (<= 10 features)   :-> **`0.5 - 1`**
       - mid dim (10 - 20 features) :-> **`1 - 2`**
       - high dim (>= 20 features)  :-> **` >2 `**

```python
# DBSCAN 

dbscan = DBSCAN(
    eps=0.5,
    min_samples=5,
)

labels = dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], c=labels)
```

```python
# DBSCAN -> change eps 

dbscan = DBSCAN(
    eps=0.18,
    min_samples=5,
)

labels = dbscan.fit_predict(X_scaled)

sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], c=labels)
```
